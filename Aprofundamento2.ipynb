```python
import pandas as pd #impotando os dados

life = pd.read_csv('https://raw.githubusercontent.com/Giseamo/aulas-paradigmas-/main/profund2/life_expectancy_years.csv')
life.head()
```

```python
life_BR = life[life.country == 'Brazil' ] #selecionando Brasil
life_BR
```

```python
life_BR = pd.melt(life_BR, id_vars=['country']) #organizando as colunas
life_BR.head()
```

```python
life_BR = life_BR.rename(columns={'variable':'year','value':'old'}) #renomeando as colunas 
life_BR.head()
```

```python
import pandas as pd #importando os dados

pobreza = pd.read_csv('https://raw.githubusercontent.com/Giseamo/aulas-paradigmas-/main/profund2/poverty_percent_people_below_550_a_day.csv')
pobreza.head()
```

```python
pobreza_BR = pobreza[ pobreza.country == 'Brazil' ] #selecionado o Brasil
pobreza_BR
```

```python
pobreza_BR1 = pd.melt(pobreza_BR, id_vars=['country']) #organizando e renomeando as colunas
pobreza_BR2 = pobreza_BR1.rename(columns={'variable':'year','value':'perc_pobreza'})
pobreza_BR2.head()
```

```python
import pandas as pd #importando os dados

saneamento = pd.read_csv('https://raw.githubusercontent.com/Giseamo/aulas-paradigmas-/main/profund2/at_least_basic_sanitation_overall_access_percent.csv')
saneamento.head()
```

```python
saneamento_BR = saneamento[ saneamento.country == 'Brazil' ] #selecionado o Brasil
saneamento_BR
```

```python
saneamento_BR1 = pd.melt(saneamento_BR, id_vars=['country']) #organizando e renomeando as colunas
saneamento_BR2 = saneamento_BR1.rename(columns={'variable':'year','value':'perc_saneamento'})
saneamento_BR2.head()
```

```python
BR = pd.merge(life_BR,pobreza_BR2,on=['country','year']) #unindo as bases
BR.head()
```

```python
BR1 = pd.merge(life_BR,saneamento_BR2,on=['country','year']) #unido as bases
BR1.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
plt.figure(figsize=(12,6))

sns.lineplot(x=BR.year, y=BR.old, label='expectativa')
sns.lineplot(x=BR.year, y=BR.perc_pobreza, label='%pobreza')


plt.title('Brasil: Expectativa X pobreza', fontsize=14)
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.lineplot(x=BR1.year, y=BR1.old, label='expectativa')
sns.lineplot(x=BR1.year, y=BR1.perc_saneamento, label='%saneamento')

plt.title('Brasil: Expectativa X saneamento', fontsize=14)
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.lineplot(x=BR.year, y=BR.old, label='expectativa')
sns.lineplot(x=BR.year, y=BR.perc_pobreza, label='%pobreza')
sns.lineplot(x=BR1.year, y=BR1.perc_saneamento, label='%saneamento')

plt.title('Brasil: Expectativa X pobreza x saneamento', fontsize=14)
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

-Qual dos indicadores tem a menor constância?
R. O indicador de pobreza, apesar de apresentar uma queda ele oscila entre alguns altos e baixos.

-Qual dos indicadores teve uma melhora mais significativa?
R.Todos os indicadores possuem uma melhora dentro do mesmo período (1999 a 2016), mas apesar de ter alguns altos e baixos o indicador de pobreza teve uma queda de 20% da população.

-É possível identificar uma correlação com a expectativa de vida e a pobreza? 
R. Não, apesar de ter uma leve semelhança na direção das linhas, a expectativa de vida não correlaciona com a pobreza, como podemos ver entre os anos de 1882 e 1886.

-É possível identificar uma correlação com a expectativa de vida e o saneamento? 
R. Sim, pois ambas as linhas caminham na mesma direção
